```python
import glob
import pandas as pd
import numpy as np
from functools import reduce
```

```python
stocks = []
for filename in np.random.choice(glob.glob('./Large_Cap/*.csv'), 10):
    stocks.append(pd.read_csv(filename, usecols = ['Symbol', 'Close Price', 'Date']))
for filename in np.random.choice(glob.glob('./Mid_Cap/*.csv'), 10):
    stocks.append(pd.read_csv(filename, usecols = ['Symbol', 'Close Price', 'Date']))
for filename in np.random.choice(glob.glob('./Small_Cap/*.csv'), 10):
    stocks.append(pd.read_csv(filename, usecols = ['Symbol', 'Close Price', 'Date']))
```

```python
stocks = [stock.rename({'Close Price': stock['Symbol'].iloc[0]}, axis = 1).drop('Symbol', axis = 1) for stock in stocks]
stocks = [stock.set_index(pd.to_datetime(stock['Date'])).drop('Date', axis = 1) for stock in stocks]
stocks = reduce(lambda x,y: pd.merge(x,y, how = 'inner', left_index = True, right_index = True), stocks)
stocks.iloc[0] == stocks.iloc[1]
```

```python
Mean_daily_return = (1 + (stocks.pct_change().mean()))**252 - 1
Std_daily_return = (stocks.pct_change().std())*(252**0.5)
```
